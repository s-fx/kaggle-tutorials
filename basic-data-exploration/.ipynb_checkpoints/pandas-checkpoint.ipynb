### DataFrame vs Series
Dataframe is a bunch of Series. <br>
DataFrame = table <br>
Series = List / single column of DataFrame <br>

```python
pd.DataFrame({"Yes": [10,12], "No": [32,10]})
```

```python
pd.DataFrame({"Person X": ["I like", "I Dont Like"], 
              "Person Y": ["OK", "BAD"]}, 
             index=["Product 1", "Product 2"])
```

```python
pd.Series([88,12,35])
```

```python
pd.Series([30, 35, 40], 
          index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

Label based selection <br>
Datas index value matters, not the position

```python
data.loc[0, "Address"]
```

```python
data.set_index("Address")
```

```python
data.Rooms == 4
```

```python
data.loc[data.Rooms == 7]
```

```python
data.loc[(data.Rooms == 7) & (data.Distance > 10)]
```

```python
data.loc[(data.Rooms == 7) | (data.Distance > 10)]
```

```python
data.loc[data.Rooms.isin([7,8])]
```

```python
data.loc[data.YearBuilt.isnull()]
```
